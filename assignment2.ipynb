## **LINEAR ALGEBTRA**

# Matrix Operations:

```python
import numpy as np
A= np.array([
    [2,-1],
    [0,3]])
B=np.array([
    [1,2],
    [-3,4]])
print("A+B=",A+B) 
print("A*B=",np.dot(A,B)) 
alpha=3
beta=2
print("3A-2B=",alpha*A-beta*B)
```

# Determinant and Inverse:

```python
import numpy.linalg as LA
A= np.array([
    [2,-1],
    [0,3]])
print("determinant=",LA.det(A)) #determinant
#since determinant is not equal to zero inverse exists so,
inverse=LA.inv(A)
print("inverse of A is=",inverse)
```

# Eigen values and Eigen vectors:

```python
C= np.array([
    [2,-1],
    [0,3]])
eigenvalues, eigenvectors= LA.eig(C)
print("C=",C)
print("Eigen value(first):")
print(eigenvalues[0])
print("Eigen value(second):")
print(eigenvalues[1])
print("Corresponding eigen vectors:")
for i in range(len(eigenvalues)):
    print(eigenvectors[:, i])
```

## Car price prediction:

```python
import numpy as np
import numpy.linalg as LA
X=np.array([
[4,1.5,100],
[6,2.0,150]
])
Y=np.array([20,30])
beta=np.dot(LA.pinv(np.dot(X.T , X)), np.dot(X.T,Y))
print("the coefficient vector for given car 1 and car 2 is givven as:",beta)
Car_third=np.array([8,3,200])
Predicted_price= np.dot(Car_third,beta)
print("The price of the third car is found to be:",Predicted_price)
```
