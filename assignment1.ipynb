## **DATA VISUALIZATION USING A HEART HEALTH DATASET**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Visualizing data:

```python
df=pd.read_csv("Health_heart_experimental.csv")
df.head(15)
```

## Line Chart:

```python

df=pd.read_csv("Health_heart_experimental.csv")
sns.lineplot(data=df, x="age", y="HR", marker="o", label="heart rate")
plt.title("Age VS Heart rate")
plt.xlabel("Age")
plt.ylabel("Heart rate(bpm)")
plt.show()
```

The above line chart shows the variation of heartrate(bpm) with age. We can clearly see here that the heart rates seems fairly stable with some noticable spike in the heart-rates among the 25 years and 47 years age groups.However, here the heartrates are ranging from(135-143 bpm) suggesting that these data may have been taken when the people were working.

## Bar Chart:

```python
sns.barplot(x='indication',y='BMI',data=df)
```

 This bar plot compares the BMI(Body mass indication) with the indication. here group 0 seems to have average BMI less than 25 while people of group 1 seem to have an average BMI of 34. Here indication 0 and 1 most likely means presence or absence of diseases. Hence we can conclude that BMI less than 25% is normal and signifies absence of diseases while BMI greater then 30 falls under unhealthy category and indicates presence of some kind of health complications.

## Pie-Chart:

```python
df['indication'].value_counts().plot(kind='pie',autopct='%1.1f%%')
```

 The given pie-plot shows the distribution of indication of presence or absence of health complications. The percentage absence of health complications in people covers 52.7% going just over half of the total. While those having health health complications covers 47.3%. The two percentages being quite near to each other signifies that in this dataset there is almost equal distribution of people having and not having health complications.

## Box-plot:

```python
sns.boxplot(df['HR'])
```

Here is a box-plot showing the variance of heart-rates(bpm). It shows here that the heart-rates are ranging from 80bpm to 200bpm with the quartiles ranging from 110bpm(25th percentile) to 170bpm(75th percentile). We can conclude from the wide range of heart-rates that the data was collected during various works or resting stages.

## Histogram:

```python
plt.hist(df['weightKg'])
```

 Here, this histogram-chart shows the distribution of people's weight(in Kg). We can see from this histogram that the number of peoples are decreasing as the weight increases with the highest number of people having weight frpm 45-50 kgs while the least number of people have the weight of about 112-117kgs. Hence we can say from this graph that higher number of people in this dataset have lower weight than those of higher weight.
